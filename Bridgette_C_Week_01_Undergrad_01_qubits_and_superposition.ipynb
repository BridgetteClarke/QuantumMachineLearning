## Overview
This notebook introduces the fundamental unit of quantum computation — the **qubit** — and its defining property of **superposition**. We explore how qubits differ from classical bits and how to represent and measure them using Qiskit.

**Learning Objectives:**
- Define a qubit and contrast it with a classical bit

- Create and measure single-qubit states in Qiskit

## 📚 Concept Review

### Classical Bit vs. Qubit
| Property | Classical Bit | Qubit |
|----------|-------------|-------|
| States | 0 or 1 | α\|0⟩ + β\|1⟩ |
| Measurement | Deterministic | Probabilistic |
| Information | 1 bit | up to 1 qubit |

### Superposition
A qubit in superposition is described by:
$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad |\alpha|^2 + |\beta|^2 = 1$$
where $|\alpha|^2$ is the probability of measuring $|0\rangle$ and $|\beta|^2$ is the probability of measuring $|1\rangle$.

## 🔧 Environment Setup

```python
# Install Qiskit if needed:
!pip install qiskit qiskit-aer

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np

sim = AerSimulator()
print('Qiskit environment ready ✓')
```

## 🔬 Section 1 — The |0⟩ State (Classical Zero)

```python
# A single qubit starts in |0⟩ by default
qc_zero = QuantumCircuit(1, 1)
qc_zero.measure(0, 0)
print('Circuit for |0⟩ state:')
print(qc_zero.draw(output='text'))

sim = AerSimulator()
job = sim.run(qc_zero, shots=1024)
counts = job.result().get_counts()
print(f'Measurement outcomes: {counts}')
```

## 🌀 Section 2 — Creating Superposition with Hadamard Gate

```python
# The Hadamard gate (H) creates equal superposition:
# H|0⟩ = (|0⟩ + |1⟩) / √2

qc_super = QuantumCircuit(1, 1)
qc_super.h(0)          # Apply Hadamard
qc_super.measure(0, 0)
print('Circuit with Hadamard gate:')
print(qc_super.draw(output='text'))
```

```python
# Simulate and plot results
job = sim.run(qc_super, shots=2048)
counts = job.result().get_counts()
print(f'Measurement outcomes: {counts}')
plot_histogram(counts, title='Superposition Measurement (H|0⟩)')
plt.show()
```

## 📐 Section 3 — Statevector Visualization

```python
# Examine the statevector BEFORE measurement
qc_sv = QuantumCircuit(1)
qc_sv.h(0)

sv = Statevector.from_instruction(qc_sv)
print(f'Statevector: {sv}')
print(f'Prob |0⟩: {abs(sv[0])**2:.4f}')
print(f'Prob |1⟩: {abs(sv[1])**2:.4f}')
```

I worked this out by hand using matrix multiplication:
H|1⟩ = (1/√2) * [[1,1],[1,-1]] * [0,1]

Top row: (1x0) + (1x1) = 1
Bottom row: (1x0) + (-1x1) = -1
Then multiply by 1/√2 → [1/√2, -1/√2]

So the probabilities are:
- P(|0⟩) = (1/√2)² = 0.50 → 50%
- P(|1⟩) = (-1/√2)² = 0.50 → 50%

The negative sign disappears when you square it, so I still expect a ~50/50 split.
It won't be exactly 50/50 because we are running a finite number of random shots —
just like flipping a coin 2048 times won't give you exactly 1024 heads.

The notebook is asking me to create a 2-qubit circuit where both qubits are in superposition
and find the probability of measuring |00⟩.

Here is my understanding of what is happening step by step:

- We have 2 qubits, both starting in |0⟩ by default — as a vector that is [1, 0]
- We apply the H gate to both qubits independently
- H puts each qubit into equal superposition — 50/50 between |0⟩ and |1⟩

Since each qubit is independently 50/50, the possible outcomes when measuring are:
- |00⟩ → qubit 0 is 0 AND qubit 1 is 0
- |01⟩ → qubit 0 is 0 AND qubit 1 is 1
- |10⟩ → qubit 0 is 1 AND qubit 1 is 0
- |11⟩ → qubit 0 is 1 AND qubit 1 is 1

That gives us 4 outcomes total — because with 2 qubits there are 2² = 4 combinations.

The probability of measuring |00⟩ is:
P(|00⟩) = P(qubit 0 is 0) x P(qubit 1 is 0) = 0.5 x 0.5 = 0.25 → 25%

Same logic applies to all four outcomes — they are all 25%.

My results confirmed this — I got roughly 512 counts for each outcome out of 2048 shots.
It wasn't exactly 512 each because measurement is random, just like flipping a coin.

## 💡 Key Takeaways
- A qubit lives in a 2D complex vector space spanned by |0⟩ and |1⟩
- The Hadamard gate creates **equal superposition** — 50/50 probability
- Measurement **collapses** the superposition to a definite state
- Probabilities are squares of amplitudes: $P(x) = |\langle x|\psi\rangle|^2$
